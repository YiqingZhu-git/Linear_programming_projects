### How does NetworkX work?
Networkx is used to study complex networks which is in form of graphs with nodes and edges. It is a resource of Python. Therefore, we need to import Networkx before using. People can generate many types of random and classic networks and analyze network structure, build, design and draw network flows.

```python
import matplotlib.pyplot as plt
import networkx as nx
```

### Create graphs and plot by using Networkx

```python
#Create a empty graph
G = nx.Graph()

#add node
G.add_node(1) #adds the node 
G.add_node(2)
G.add_node(3)
G.add_node(4)

G.add_edge(1,2) #adds the edges
G.add_edge(1,3)
G.add_edge(1,4)
G.add_edge(2,4)
G.add_edge(3,1)
G.add_edge(3,4)
```

```python
G.nodes()
```

```python
nx.draw(G, with_labels = 1, node_color = "red" ) #label the graph from the number 1
```

Draw the G with a circular layout.

```python
nx.draw_circular(G)
```

Draw the path by 20 nodes

```python
Q = nx.path_graph(20)
nx.draw(Q, with_labels = 1, node_color='green')
```

We can also use Networkx to simuate the network model. For example, the Barabasi Albert graph on 200 nodes.

```python
T = nx.barabasi_albert_graph(200, m=2 , seed = 1)
nx.draw(T,pos=nx.spring_layout(T))
```

Promote the graph T by using matplotlib and Networkx

```python
def PromoteGraph(graph):
    #Set the figure size
    plt.figure(figsize = (10,10))
    #set the position
    pos = nx.kamada_kawai_layout(graph)
    #set the node and edge
    node_appearance = {"node_color" : "orange", "node_size" : 30}
    edge_appearance = {"edge_color" : "green", "width" : 0.5, "alpha" :0.3}
    nx.draw_networkx_edges(graph, pos, **edge_appearance)
    nx.draw_networkx_nodes(graph, pos, **node_appearance)

    plt.show(graph)
```

```python
PromoteGraph(T)
```

```python
#promote for the other graph
E = nx.dodecahedral_graph()
nx.draw(E, pos=nx.spring_layout(E)) 
PromoteGraph(E)
```
